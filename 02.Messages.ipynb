### 2.4 Tool Message

### æ¶ˆæ¯
æˆ‘ä»¬æ¥åˆ›å»ºä¸€ä¸ªå·¥å…·ï¼Œè®©Agentäº§å‡ºä¸€äº›å·¥å…·æ¶ˆæ¯ã€‚

````python
from langchain.tools import tool
@tool
def check_haiku_lines(haiku: str) -> str:
    """æ£€æŸ¥ç»™å®šçš„ä¿³å¥æ˜¯å¦æ­£å¥½æœ‰ 3 è¡Œã€‚
    å¦‚æœæ­£ç¡®åˆ™è¿”å› None,å¦åˆ™è¿”å›é”™è¯¯ä¿¡æ¯ã€‚
    """
    # å°†æ–‡æœ¬æ‹†åˆ†ä¸ºå¤šè¡Œï¼Œå¿½ç•¥é¦–å°¾ç©ºæ ¼
    lines = [line.strip() for line in haiku.strip().splitlines() if line.strip()]
    print(f"æ£€æŸ¥ä¿³å¥ï¼Œå®ƒæœ‰ {len(lines)} è¡Œ:\n {haiku}")

    if len(lines) != 3:
        return f"é”™è¯¯ï¼è¿™é¦–ä¿³å¥æœ‰ {len(lines)} è¡Œã€‚ä¿³å¥å¿…é¡»æ°å¥½æœ‰ 3 è¡Œã€‚"
    return "æ­£ç¡®!è¿™é¦–ä¿³å¥æœ‰3è¡Œã€‚"

agent = create_agent(
    model=model,
    tools=[check_haiku_lines],
    system_prompt="ä½ æ˜¯ä¸€ä¸ªä½“è‚²è¯—æ­Œä¸“å®¶ï¼Œä¼šè¡¥å…¨å¼€å¤´çš„ä¿³å¥ã€‚ä½ å¿…é¡»è°ƒç”¨å·¥å…·æ£€æŸ¥ä½ çš„ä½œå“ã€‚",
)

result = agent.invoke({"messages": "è¯·ä¸ºæˆ‘å†™ä¸€é¦–å…³äºç¯®çƒè¿åŠ¨å‘˜çš„ä¿³å¥ã€‚"})
````
